### Derivation of Kinematics Equations
<hr>

What exactly is kinematics? Kinematics is a branch of physics that studies the motion of objects without considering the forces that cause the motion. It is a field that emphasizes the relationships between the positions, velocity, and acceleration of moving objects. Equations that describe the motion of a projectile (kinematics) can be derived from the definitions of average acceleration and average velocity.

$$\boxed{\text{(a) Average Acceleration = } \bar{a} = \frac{\Delta v}{\Delta t} = \frac{v_2 - v_1}{t_2 - t_1}}$$
$$\boxed{\text{(b) Average Velocity = } \bar{v} = \frac{\Delta x}{\Delta t} = \frac{x_2 - x_1}{t_2 - t_1} = \frac{v_0 + v}{2}}$$

<div style="display:flex; justify-content:center; gap:10px;">
    
<span style="display:inline-block;">Where: $v_1$ = starting velocity,</span>
<span style="display:inline-block;"> $v_2$ = final velocity,</span>
<span style="display:inline-block;"> $x_1$ = starting position,</span>
<span style="display:inline-block;"> $x_2$ = final position,</span>
<span style="display:inline-block;"> $t_1$ = initial time,</span>
<span style="display:inline-block;">and  $t_2$ = end time.</span>

</div>

In the above, $velocity$ and $time$ are used to describe $acceleration$, while $position$ and $time$ describe $velocity$. In this simulation, the initial time will always be equal to zero, meaning $t_1 = 0$. By substituting this into equation (1), and solving for $v$, you can get the first kinematics equation:

\begin{aligned}
(1)\quad & \boxed{v = v_0 + at}
\end{aligned}

Deriving $ \bar{v} = \frac{v_0 + v}{2} $ from equation (b) can be accomplished using geometry or integration. From integration, start with equation (1): $v(t) = v_0 + at$, and integrate both sides of the equation with respect to time $t$ from $t_1$ to $t_2$. This gives $\int_{t_1}^{t_2} v dt = \int_{t_1}^{t_2} (v_0 + at)dt \rightarrow v(t) \left. \right|_{t_1}^{t_2} = v_0t \left. \right|_{t_1}^{t_2} + \frac{a t^2}{2}\left. \right|_{t_1}^{t_2} \rightarrow \bar{v} = v_0 + \frac{1}{2}a(t)$. Equation (a) can be rewritten as $at = v - v_0$ and plugged into equation (1) to get $\bar{v} = v_0 + \frac{1}{2}(v-v_0) \rightarrow \boxed{\bar{v} = \frac{v_0 + v}{2}}$. Using equation (b), which was just derived, and the relation $\Delta{x} = \bar{v}t$, you get the second kinematic equation 

\begin{aligned}
(2)\quad & \boxed{\Delta{x} = \frac{1}{2}(v_0+v)t}
\end{aligned}

Next, substituting the expression for $v$ from the equation (1) into equation (2), you can get the third kinematics equation. 

\begin{aligned}
(3)\quad & \boxed{\Delta{x} = v_0t + \frac{1}{2}at^2}
\end{aligned}

Finally, if you take equation (1), solve for $t$, and substitute $t$ into equation (2) and solve for $v^2$, you obtain the fourth kinematics equation.

\begin{aligned}
(4)\quad & \boxed{v^2 = v_0^2 + 2a\Delta{x}}
\end{aligned}

The four kinematics equations that can be used to decribe the motion of a projectile have been derived. In this simulation these equations and knowledge with vector math will be used to model projectile motion. 

### Understanding The Drag Force
<hr>

Although these equations assume motion with constant acceleration, this simulation uses numerical integration because drag makes acceleration velocity-dependent. This means that drag acts as a force on projectiles and slows it down as it moves through the air. If this wasn't accounted, the projectiles would land much farther than you would see in a real-life demonstration. In order to account for drag, it is helpful to understand the various types of drag to most accurately account for it. Drag can take on three "classical" forms. Note: There are more than three types of drag such as drag on a spinning bullet (Magnus Effect) or drag that is wind-relative. This simulation will assume the projectile that is being launches has no rotation, no wind speed, etc.

* ${Linear\text{ }Drag:} \text{ }(F_d = -bv)$ <br>
  Best for small, slower moving objects where flow remains smooth.<br><br>
* ${Quadratic\text{ }Drag:} \text{ }(F_d = -cv^2)$ <br>
  Most accurate for larger, faster objects (e.g. baseballs, arrows, bullets) where flow is turbulent.<br><br>
* ${Mixed\text{ }Drag:} \text{ }(F_d = -bv - cv^2)$ <br>
  Applies when an object starts fast (turbulent flow) but slow enough for the flow to become smooth.

In order to determine if a medium's flow is laminar (smooth) or turbulent (chaotic), you measure the ratio of inertial forces to viscous forces in a fluid. This ratio is known as the Reynolds Number and can be calculated using.

\begin{aligned}
\quad & \boxed{R_e = \frac{uL}{v}}
\end{aligned}

The Reynolds Number is dependent on: $u$ = flow velocity ($\frac{m}{s}$), $L$ = characteristic length ($m$), $v$ = kinematic viscosity ($\frac{m}{s}$).
        
For now, do not worry about where $b$ or $c$ originate from, I will cover their derivation in the next section. In the simulation, the initial properties of the projectile will match a baseball. A baseball is large compared to air molecules and travels fast enough that turbulent flow dominates. With this in mind, a quadratic drag model will be most accurate to describe it's motion. Including gravity and quadratic dra, the net force on the projectile is: $$\boxed{\sum\vec{F} = m\vec{g} - c{v^2}{\hat{v}}}$$

You can break up the net force on the projectile into its x and y components using the velocity unit vector, $\hat{v} = \frac{\vec{v}}{||v||}$, where $||v|| = \sqrt{v_x^2 + v_y^2} = v$. Plugging the new velocities into the equation for the forces on the projectile, you end up with the x and y components describing the projectile's acceleration:

\begin{aligned}
\; & \boxed{a_x = -\frac{c}{m} v v_x}
\quad \text{and} \quad
\; \boxed{a_y = g_y - \frac{c}{m} v v_y}
\end{aligned}

### Explaining Bernoulli's Equation
<hr>

When we covered drag there were variables such as $v$ and $c$. In the case of $c$ this comes from the equation $F_d = \frac{1}{2}\rho{C_d}A{v^2}$. This equation is derived from fluid mechanics concepts starting with Bernoulli's Equation.

\begin{aligned}
(5)\quad & \boxed{p_1 + \frac{1}{2}\rho{v_1^2} + \rho{g}{h_1} = p_2 + \frac{1}{2}\rho{v_2^2} + \rho{g}{h_2}}
\end{aligned}

<div style="display:flex; justify-content:center; gap:15px;">
    
<span style="display:inline-block;">Where: $p_{1,2}$ = static pressure,</span>
<span style="display:inline-block;"> $\rho$ = density of the fluid,</span>
<span style="display:inline-block;"> $v_{1,2}$ = flow speed of fluid,</span>
<span style="display:inline-block;"> $g$ = gravitational acceleration,</span>
<span style="display:inline-block;">and $h_{1,2}$ = reference height.</span>
</div>

When both heights are the same, $\rho$, $g$, and $h$ cancels out. Using equation (5) to solve for $p_1$ you get the equation $p_1 = p_2 + \frac{1}{2}\rho{v^2}$. If you arrange it so that both static pressure is on one side of the equation you get $p_1 - p_2 = \frac{1}{2}\rho{v^2}$. Here, $p_1 - p_2$ is something called dynamic pressure which I will represent with $q$. Dynamic pressure is the kinetic energy per unit volume inside a moving fluid. It's formula is written as $q = \frac{1}{2}\rho{v^2}$. The relationship $Force$ = $Pressure$ $*$ $Area$ can be used to convert dynamic pressure into a force. Doing so gives $F_p = qA \rightarrow F_p = (\frac{1}{2}\rho{v^2})A$. This gives the drag force for an object that is flat. In order to get the drag force that accounts for an objects shape, you need to introduce a variable that is dimensionless that is known as a correction factor $C_d$. Doing so, this gives the final drag equation:

\begin{aligned}
(6)\quad & \boxed{F_d = \frac{1}{2}\rho{C_d}{A}{v^2}}
\end{aligned}

To make things simpler you can use the lumped drag constant $c$, which has units $\frac{kg}{m}$:

\begin{aligned}
(7)\quad & \boxed{c = \frac{1}{2} \rho C_d A} 
\end{aligned}

This simulation will use the lumped drag constant of various objects to model drag on various projectiles.
You have covered all of the math needed in our simulation (and then some). You can now move from the theoretical physics to computational physics. Note: For the simulation of a projectile that has no drag, the mass will not affect things. Therefore, you'll notice that selecting various presets does not affect the impact distance and impact speed.

```python
%matplotlib inline

%matplotlib ipympl

#Projectile Motion with Drag. Author: Zachary Lee
#--------------------------------------------------------------------------
#In the past, I have used Euler's method to handle updates/timesteps.
#For this simulation I will be using scipy to handle the integration.
#Due to the increase in accuracy and computational efficiency.
#--------------------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
import pandas as pd
from scipy.integrate import solve_ivp
from matplotlib.animation import FuncAnimation

plt.rcParams['animation.embed_limit'] = 50 # Increases memory limit for animations to 50 MB.

g = 9.81
max_time = 30
time_step = 0.01

SimConfig = {"t_step": (0.0, max_time), "time_step": time_step, "dense_output": True, "rtol": 1e-6, "atol": 1e-8}

#---------------------------------[Functions]---------------------------------------

def kinematics_drag(t, components, m, c):         #Handles the components for a projectile with drag.
    x, y, v_x, v_y = components                   #Array to store components.
    v = np.sqrt(v_x**2 + v_y**2)                  #Calculates speed.
    a_x = -(c/m) * v_x * v                        #Quadratic Drag x_component.
    a_y = -g - (c/m) * v_y * v                    #Quadratic Drag y_component.
    return [v_x, v_y, a_x, a_y]                   #Returns component values.
    
def kinematics_nd(t, components, m, c):           #Handles the components for a projectile without drag.
    x_nd, y_nd, v_x_nd, v_y_nd = components
    return [v_x_nd, v_y_nd, 0, -g]
   
def impact(t, components, m, c):                  #Stops the simulation when y=0 (projectile impacts ground).
    return components[1]
       
impact.terminal = True                            #Stops integration when y=0.
impact.direction = -1                             #Triggers the event when y changes in the downward direction.
                                                  #Without it, the event would trigger at the beginning of the simulation.

def update_simulation(m=0.15, c = 0.00086, y_0 = 0, angle = 45, v_0 = 30, display_results=True):
    
    #Statements to handle possible errors from unrealistic values.
    if m <= 0: raise ValueError("Mass must be positive.")
    if c < 0: raise ValueError("Lumped Drag Coefficient must be positive.")
    if v_0 < 0: raise ValueError("Speed cannot be negative.")
    if not (0 <= angle <= 90): raise ValueError("Launch angle must be between 0 and 90.")
    if y_0 < 0: raise ValueError("Initial Height cannot be negative.")
       
    impact_values.clear_output() #Clears the previously displayed impact value.
    theta = np.radians(angle) #Converts the angle from radians to degrees.
   
    v_x0 = v_0 * np.cos(theta) #Vector components for the intial velocity.
    v_y0 = v_0 * np.sin(theta)
    components_0 = [0, y_0, v_x0, v_y0] #Array to store projectile's initial values.
   
                                                  #Numerically solves both sets of kinematic equations for the projectile.
    solution = solve_ivp(kinematics_drag, t_span = (0, max_time), y0 = components_0, args = (m, c), events = impact, max_step = SimConfig["time_step"], dense_output=SimConfig["dense_output"], rtol=SimConfig["rtol"], atol=SimConfig["atol"])
    solution_nd = solve_ivp(kinematics_nd, t_span = (0, max_time), y0 = components_0, args = (m, c), events = impact, max_step = SimConfig["time_step"], dense_output=SimConfig["dense_output"], rtol=SimConfig["rtol"], atol=SimConfig["atol"])
   
    t = solution.t #Values for projectile (drag).
    x, y, v_x, v_y = solution.y
    v = np.sqrt(v_x**2 + v_y**2)
   
    t_nd = solution_nd.t #Values for projectile (no drag).
    x_nd, y_nd, v_x_nd, v_y_nd = solution_nd.y
    v_nd = np.sqrt(v_x_nd**2 + v_y_nd**2)
  
    KE = 0.5 * m * v**2 #Values for Impact Energy of projectile (drag).
    PE = m * g * y
    E_t = KE + PE
   
    KE_nd = 0.5 * m * v_nd**2 #Values for Impact Energy of projectile (no drag).
    PE_nd = m * g * y_nd
    E_t_nd = KE_nd + PE_nd
  
    x_impact = x[-1] #Stores values of projectile (drag) at impact.
    v_impact = v[-1]
    E_impact = E_t[-1]
    x_nd_impact = x_nd[-1] #Gets values of projectile (no drag) at impact.
    v_nd_impact = v_nd[-1]
    E_nd_impact = E_t_nd[-1]

    if display_results:
        with impact_values: #Prints both sets of value of the projectiles.
            print(f"Impact distance (drag):      {x_impact:.2f} m")
            print(f"Impact distance (no drag):   {x_nd_impact:.2f} m")
            print(f"Distance Difference:         {abs(x_impact - x_nd_impact):.2f} m")
            print(f"Percent Error (no drag):     {(abs(x_impact - x_nd_impact)/x_impact)*100:.2f} %")
            print("---------------------------------------------------------------")
            print(f"Impact Speed (drag):         {v_impact:.2f} m/s")
            print(f"Impact Speed (no drag):      {v_nd_impact:.2f} m/s")
            print(f"Speed Difference:            {abs(v_impact - v_nd_impact):.2f} m/s")
            print(f"Percent Error (no drag):     {(abs(v_impact - v_nd_impact)/v_impact)*100:.2f} %")
            print("---------------------------------------------------------------")
            print(f"Impact Energy (drag):        {E_impact:.2f} joules")
            print(f"Impact Energy (no drag):     {E_nd_impact:.2f} joules")
            print(f"Energy Difference:           {abs(E_impact - E_nd_impact):.2f} joules")
            print(f"Percent Error (no drag):     {(abs(E_impact - E_nd_impact)/E_impact)*100:.2f} %")
  
    # Returns both sets of values of the projectiles to be used with the "run" button.
    results_simulation = {"t": t, "x": x, "y": y, "v": v, "E_t": E_t, 
                          "t_nd": t_nd, "x_nd": x_nd, "y_nd": y_nd, "v_nd": v_nd, "E_t_nd": E_t_nd}
    return results_simulation
       
def plot_simulation(results, show_nodrag_path=True):
 
    plt.close('all')
 
    t = results["t"]
    x = results["x"]
    y = results["y"]
  
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))  # Single plot only

    # Plots the static trajectories.
    ax.plot(x, y, label="With Drag", color="blue", linestyle="--")

    if show_nodrag_path:
        ax.plot(results["x_nd"], results["y_nd"], label="Without Drag", color="red", linestyle="--")
   
    # Animation elements.
    point_d, = ax.plot([], [], 'o', color='blue', markersize=11, label="Projectile (with drag)")
    trail_d, = ax.plot([], [], color='blue', alpha=0.6, linewidth=3.5) #Creates a trail that fades.
   
    ax.set_xlabel("x (meters)")
    ax.set_ylabel("y (meters)")
    ax.set_title("Projectile Motion: Drag vs No Drag")
    ax.grid(True)
    ax.legend()
 
    plt.tight_layout()
  
    # Creates an animation for the projectile.
    def init():
        point_d.set_data([], [])
        trail_d.set_data([], [])
        return point_d, trail_d
   
    def animate(frame):
        idx = min(frame, len(x)-1)
        point_d.set_data([x[idx]], [y[idx]])
        trail_len = min(25, idx+1)
        trail_d.set_data(x[idx-trail_len+1:idx+1], y[idx-trail_len+1:idx+1])
        return point_d, trail_d
   
    # Controls how many frames to generate for the animation. Note: more frames = longer time for plot to generate.
    n_frames = len(t)
    step = max(1, n_frames // 120)   
    frames = range(0, n_frames, step)
   
    anim = FuncAnimation(fig, animate, init_func=init, frames=frames, interval=45, blit=False, repeat=True)
   
    # Keep reference
    fig.animation = anim
   
    # Reliable display method for Jupyter
    from IPython.display import HTML
    html_anim = HTML(anim.to_jshtml())
   
    return fig, html_anim, anim
    
#Creates sliders to adjust initial values of the projectile.
m_slider = widgets.FloatSlider(value=0.145, min=0.01, max=20, step=0.001, description="Mass (kg)")
c_slider = widgets.FloatSlider(value=0.00086, min=0.0, max=0.005, step=0.0001, description="Lumped Drag Coefficient", readout_format=".5f")
y_0_slider = widgets.FloatSlider(value=0, min=0, max=100, step=1, description="Initial Height (m)")
angle_slider = widgets.FloatSlider(value=45, min=0, max=90, step=1, description="Angle (degrees)")
v_0_slider = widgets.FloatSlider(value=30, min=1, max=100, step=1, description="Initial Speed (m/s)")

#Creates a button to reset the parameters to a baseball.
reset_button = widgets.Button(description="Reset", button_style="warning", tooltip = "Reset all sliders to inital values")
run_button = widgets.Button(description="Run", button_style="primary", tooltip = "Run the simulation")

#Creates presets of various projectiles using real-world mass and lumped drag constant values.
projectile_preset = widgets.Dropdown(options = {"Ping-Pong Ball (Mass: = 0.0027 kg, LC_d: = 0.000362 kg/m)": (0.0027, 0.000362), "Baseball (Mass: 0.145 kg, LC_d: = 0.00084 kg/m)": (0.145, 0.00084),
                                                "Soccerball (Mass: = 0.450 kg, LC_d: = 0.00582 kg/m)": (0.450, 0.00582), "Coconut (Mass: 2.00 kg, LC_d: = 0.00701 kg/m)": (2.00, 0.00701),
                                                "Shotput (Mass: = 7.26 kg, LC_d = 0.00326 kg/m)": (7.26, 0.00324), "Large Cannonball (Mass: = 14.51 kg, LC_d: = 0.00443 kg/m)": (14.51, 0.00443)
                                               },
                                                value = (0.145, 0.00084),
                                                description = "Projectile Presets"
                                    )
def update_slider(update):
    mass, c = update['new']
    m_slider.value = mass
    c_slider.value = c
    
projectile_preset.observe(update_slider, names="value")

def run_and_plot(b):
    error_output.clear_output()
    plot_output.clear_output(wait=True)
    try:
        results_simulation = update_simulation(m_slider.value, c_slider.value, y_0_slider.value, angle_slider.value, v_0_slider.value)
        with plot_output:
            plot_output.clear_output(wait=True)
            fig, html_anim, anim = plot_simulation(results_simulation, nodrag_path_toggle.value)
            display(html_anim)
            
    except ValueError as e:
        with error_output:
            print(f"Input error: {e}")
            
    except Exception as e:
        with error_output:
            print(f"Unexpected error: {e}")
    
run_button.on_click(run_and_plot)

#Function that holds the values that the reset button, when clicked, sets each one to.
def reset_slider(b):
   
    m_slider.value = 0.145
    c_slider.value = 0.00084
    y_0_slider.value = 0
    angle_slider.value = 45
    v_0_slider.value = 30
    projectile_preset.value = (0.145, 0.00084)
    
reset_button.on_click(reset_slider)

impact_values = widgets.Output()
plot_output = widgets.Output()
error_output = widgets.Output()

nodrag_path_toggle = widgets.Checkbox(value=True, description="Show No-Drag Line", indent=False)
save_gif_button = widgets.Button(description="Save Animation as GIF", button_style="success", tooltip="Save current animation as GIF")
export_csv_button = widgets.Button(description="Export Trajectory to CSV", button_style="info", tooltip="Export Data to CSV")
sweep_button = widgets.Button(description="Run Angle Sweep", button_style="primary", tooltip="Show range vs. launch angle")

def save_gif(b):
    try:
        results = update_simulation(m_slider.value, c_slider.value, y_0_slider.value, angle_slider.value, v_0_slider.value)
        _, _, anim = plot_simulation(results, nodrag_path_toggle.value)
        filename = f"projectile_angle_{int(angle_slider.value)}deg.gif"
        anim.save(filename, writer='pillow', fps=30)
        print(f"Saved GIF: {filename}")
        
    except Exception as e:
        print(f"Error saving GIF: {e}")

def export_csv(b):
    try:
        results = update_simulation(m_slider.value, c_slider.value, y_0_slider.value, angle_slider.value, v_0_slider.value)
        df = pd.DataFrame({
            'Time (s)': results['t'],
            'x (m)': results['x'],
            'y (m)': results['y'],
            'Speed (m/s)': results['v']
        })
        filename = f"projectile_angle_{int(angle_slider.value)}deg.csv"
        df.to_csv(filename, index=False)
        print(f"Saved: {filename}")
    except Exception as e:
        print(f"Error exporting CSV: {e}")

def run_sweep(b):
    try:
        angles = np.arange(10, 85, 5)
        ranges_drag = []
        ranges_nd = []

        for ang in angles:
            res = update_simulation(m_slider.value, c_slider.value, y_0_slider.value, ang, v_0_slider.value, display_results=False)
            ranges_drag.append(res['x'][-1])
            ranges_nd.append(res['x_nd'][-1])

        plt.figure(figsize=(9, 6))
        plt.plot(angles, ranges_drag, 'b-o', label="With Drag")
        plt.plot(angles, ranges_nd, 'r--o', label="No Drag")
        plt.xlabel("Launch Angle (°)")
        plt.ylabel("Range (meters)")
        plt.title("Projectile Range vs Launch Angle")
        plt.grid(True)
        plt.legend()
        plt.show()

    except Exception as e:
        print("Sweep Error:",e)


save_gif_button.on_click(save_gif)
export_csv_button.on_click(export_csv)
sweep_button.on_click(run_sweep)

#Stacks the reset button in a vertical column
ui = widgets.VBox([impact_values, error_output, m_slider, c_slider, y_0_slider, angle_slider, v_0_slider, projectile_preset, nodrag_path_toggle, widgets.HBox([run_button, reset_button, sweep_button, save_gif_button, export_csv_button]), plot_output])

#Shows everthing on the screen.
display(ui)
```
