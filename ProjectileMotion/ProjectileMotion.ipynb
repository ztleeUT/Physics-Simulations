### Derivation of Kinematics Equations
<hr>

What exactly is kinematics? Kinematics is a branch of physics that studies the motion of objects without considering the forces that cause the motion. It is a field that emphasizes the relationships between the positions, velocity, and acceleration of moving objects. Equations that describe the motion of a projectile (kinematics) can be derived from the definitions of average acceleration and average velocity.

$$\boxed{\text{(a) Average Acceleration = } \bar{a} = \frac{\Delta v}{\Delta t} = \frac{v_2 - v_1}{t_2 - t_1}}$$
$$\boxed{\text{(b) Average Velocity = } \bar{v} = \frac{\Delta x}{\Delta t} = \frac{x_2 - x_1}{t_2 - t_1} = \frac{v_0 + v}{2}}$$

<div style="display:flex; justify-content:center; gap:10px;">
    
<span style="display:inline-block;">Where: $v_1$ = starting velocity,</span>
<span style="display:inline-block;"> $v_2$ = final velocity,</span>
<span style="display:inline-block;"> $x_1$ = starting position,</span>
<span style="display:inline-block;"> $x_2$ = final position,</span>
<span style="display:inline-block;"> $t_1$ = initial time,</span>
<span style="display:inline-block;">and  $t_2$ = end time.</span>

</div>

In the above, $velocity$ and $time$ are used to describe $acceleration$, while $position$ and $time$ describe $velocity$. In this simulation, the initial time will always be equal to zero, meaning $t_1 = 0$. By substituting this into equation (1), and solving for $v$, you can get the first kinematics equation:

\begin{aligned}
(1)\quad & \boxed{v = v_0 + at}
\end{aligned}

Deriving $ \bar{v} = \frac{v_0 + v}{2} $ from equation (b) can be accomplished using geometry or integration. From integration, start with equation (1): $v(t) = v_0 + at$, and integrate both sides of the equation with respect to time $t$ from $t_1$ to $t_2$. This gives $\int_{t_1}^{t_2} v dt = \int_{t_1}^{t_2} (v_0 + at)dt \rightarrow v(t) \left. \right|_{t_1}^{t_2} = v_0t \left. \right|_{t_1}^{t_2} + \frac{a t^2}{2}\left. \right|_{t_1}^{t_2} \rightarrow \bar{v} = v_0 + \frac{1}{2}a(t)$. Equation (a) can be rewritten as $at = v - v_0$ and plugged into equation (1) to get $\bar{v} = v_0 + \frac{1}{2}(v-v_0) \rightarrow \boxed{\bar{v} = \frac{v_0 + v}{2}}$. Using equation (b), which was just derived, and the relation $\Delta{x} = \bar{v}t$, you get the second kinematic equation 

\begin{aligned}
(2)\quad & \boxed{\Delta{x} = \frac{1}{2}(v_0+v)t}
\end{aligned}

Next, substituting the expression for $v$ from the equation (1) into equation (2), you can get the third kinematics equation. 

\begin{aligned}
(3)\quad & \boxed{\Delta{x} = v_0t + \frac{1}{2}at^2}
\end{aligned}

Finally, if you take equation (1), solve for $t$, and substitute $t$ into equation (2) and solve for $v^2$, you obtain the fourth kinematics equation.

\begin{aligned}
(4)\quad & \boxed{v^2 = v_0^2 + 2a\Delta{x}}
\end{aligned}

The four kinematics equations that can be used to decribe the motion of a projectile have been derived. In this simulation these equations and knowledge with vector math will be used to model projectile motion. 

### Understanding The Drag Force
<hr>

Although these equations assume motion with constant acceleration, this simulation uses numerical integration because drag makes acceleration velocity-dependent. This means that drag acts as a force on projectiles and slows it down as it moves through the air. If this wasn't accounted, the projectiles would land much farther than you would see in a real-life demonstration. In order to account for drag, it is helpful to understand the various types of drag to most accurately account for it. Drag can take on three "classical" forms. Note: There are more than three types of drag such as drag on a spinning bullet or drag that is wind-relative. This simulation will cover the three that are most relevant. 

* ${Linear\text{ }Drag:} \text{ }(F_d = -bv)$ <br>
  Best for small, slower moving objects where flow remains smooth.<br><br>
* ${Quadratic\text{ }Drag:} \text{ }(F_d = -cv^2)$ <br>
  Most accurate for larger, faster objects (e.g. baseballs, arrows, bullets) where flow is turbulent.<br><br>
* ${Mixed\text{ }Drag:} \text{ }(F_d = -bv - cv^2)$ <br>
  Applies when an object starts fast (turbulent flow) but slow enough for the flow to become smooth.

        
For now, do not worry about where $b$ or $c$ originate from, I will cover their derivation in the next section. In the simulation, the initial properties of the projectile will match a baseball. A baseball is large compared to air molecules and travels fast enough that turbulent flow dominates. With this in mind, a quadratic drag model will be most accurate to describe it's motion. Including gravity and quadratic dra, the net force on the projectile is: $$\boxed{\sum\vec{F} = m\vec{g} - c{v^2}{\hat{v}}}$$

You can break up the net force on the projectile into its x and y components using the velocity unit vector, $\hat{v} = \frac{\vec{v}}{||v||}$, where $||v|| = \sqrt{v_x^2 + v_y^2} = v$. Plugging the new velocities into the equation for the forces on the projectile, you end up with the x and y components describing the projectile's acceleration:

\begin{aligned}
\; & \boxed{a_x = -\frac{c}{m} v v_x}
\quad \text{and} \quad
\; \boxed{a_y = g_y - \frac{c}{m} v v_y}
\end{aligned}

### Explaining Bernoulli's Equation
<hr>

When we covered drag there were variables such as $v$ and $c$. In the case of $c$ this comes from the equation $F_d = \frac{1}{2}\rho{C_d}A{v^2}$. This equation is derived from fluid mechanics concepts starting with Bernoulli's Equation.

\begin{aligned}
(5)\quad & \boxed{p_1 + \frac{1}{2}\rho{v_1^2} + \rho{g}{h_1} = p_2 + \frac{1}{2}\rho{v_2^2} + \rho{g}{h_2}}
\end{aligned}

<div style="display:flex; justify-content:center; gap:15px;">
    
<span style="display:inline-block;">Where: $p_{1,2}$ = static pressure,</span>
<span style="display:inline-block;"> $\rho$ = density of the fluid,</span>
<span style="display:inline-block;"> $v_{1,2}$ = flow speed of fluid,</span>
<span style="display:inline-block;"> $g$ = gravitational acceleration,</span>
<span style="display:inline-block;">and $h_{1,2}$ = reference height.</span>
</div>

When both heights are the same, $\rho$, $g$, and $h$ cancels out. Using equation (5) to solve for $p_1$ you get the equation $p_1 = p_2 + \frac{1}{2}\rho{v^2}$. If you arrange it so that both static pressure is on one side of the equation you get $p_1 - p_2 = \frac{1}{2}\rho{v^2}$. Here, $p_1 - p_2$ is something called dynamic pressure which I will represent with $q$. Dynamic pressure is the kinetic energy per unit volume inside a moving fluid. It's formula is written as $q = \frac{1}{2}\rho{v^2}$. The relationship $Force$ = $Pressure$ $*$ $Area$ can be used to convert dynamic pressure into a force. Doing so gives $F_p = qA \rightarrow F_p = (\frac{1}{2}\rho{v^2})A$. This gives the drag force for an object that is flat. In order to get the drag force that accounts for an objects shape, you need to introduce a variable that is dimensionless that is known as a correction factor $C_d$. Doing so, this gives the final drag equation:

\begin{aligned}
(6)\quad & \boxed{F_d = \frac{1}{2}\rho{C_d}{A}{v^2}}
\end{aligned}

To make things simpler you can use the lumped drag constant $c$, which has units $\frac{kg}{m}$:

\begin{aligned}
(7)\quad & \boxed{c = \frac{1}{2} \rho C_d A} 
\end{aligned}

This simulation will use the lumped drag constant of various objects to model drag on various projectiles.
You have covered all of the math needed in our simulation (and then some). You can now move from the theoretical physics to computational physics. Note: For the simulation of a projectile that has no drag, the mass will not affect things. Therefore, you'll notice that selecting various presets does not affect the impact distance and impact speed.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
from scipy.integrate import solve_ivp #In the past, I have used Euler's method to handle updates/timesteps. After learning of scipy, I decided to use it to handle them due to it's accuracy and computational efficiency. 

g = 9.81

#Creates a function that handles the projectile when drag is applied to it.
def kinematics_drag(t, components, m, c):
    x, y, v_x, v_y = components #Stores the projectiles position and velocity in an array.
    v = np.sqrt(v_x**2 + v_y**2) #Uses the velocity x and y components to get its speed.
    a_x = -(c/m) * v_x * v #Acceleration x component with drag.
    a_y = -g - (c/m) * v_y * v #Acceleration y component with gravity and drag.
    return [v_x, v_y, a_x, a_y] #Returns the x and y components of the projectile's velocity and acceleration.
    
def kinematics_nd(t, components, m, c): #Does the same thing as "kinematics_drag" from above for the projectile without drag.
    x_nd, y_nd, v_xnd, v_ynd = components
    return [v_xnd, v_ynd, 0, -g]
    
def impact(t, components, m, c): #function from Scipy that stops the simulation when y=0 (projectile hits the ground).
    if t < 1e-6: #added if statement to prevent code from freezing.
        return 1
    return components[1]
        
impact.terminal = True #stops integration when y=0.
impact.direction = -1 #triggers the event when y changes in the downward direction. Without it, the event would trigger at the beginning of the simulation.

#Handles the calculations using SciPy. Included are kinematics, speed vs. time, and Energy calculations of the projectile.
def update_simulation(m=0.15, c = 0.00086, y_0 = 0, angle = 45, v_0 = 30, create_png=False):

    if create_png == False: #used to with code below to create a .png file containing plots.
        plt.close('all')

    impact_values.clear_output()

    theta = np.radians(angle)
    v_x0 = v_0 * np.cos(theta)
    v_y0 = v_0 * np.sin(theta)
    components_0 = [0, y_0, v_x0, v_y0]

    solution = solve_ivp(kinematics_drag, t_span = (0, 30), y0 = components_0, args = (m, c), events = impact, max_step = 0.01, dense_output = True)
    solution_nd = solve_ivp(kinematics_nd, t_span = (0,30), y0 = components_0, args = (m, c), events = impact, max_step = 0.01, dense_output = True)

    #Used to calculate the speed of the projectile (with drag) throughout it's flight.
    t = solution.t
    x, y, v_x, v_y = solution.y
    v = np.sqrt(v_x**2 + v_y**2)

    #Used to calculate the speed of the projectile (no drag) throughout it's flight.
    t_nd = solution_nd.t
    x_nd, y_nd, v_xnd, v_ynd = solution_nd.y
    v_nd = np.sqrt(v_xnd**2 + v_ynd**2)

    #Calculates the Kinetic and Potential Energy of the projectile (with drag) to use in a plot below.
    KE = 0.5 * m * v**2
    PE = m * g * y
    E_t = KE + PE

    #Calculates the Kinetic and Potential Energy of the projectile (without drag) to use in a plot below.
    KE_nd = 0.5 * m * v_nd**2
    PE_nd = m * g * y_nd
    E_tnd = KE_nd + PE_nd
    
    #Gets the value of the position, speed, and energy of the projectile (with and without drag) when it impacts the ground.
    x_impact = x[-1]
    v_impact = v[-1]
    E_impact = E_t[-1]

    x_nd_impact = x_nd[-1]
    v_nd_impact = v_nd[-1]
    E_nd_impact = E_tnd[-1]

    with impact_values:
        print(f"Impact distance (with drag): {x_impact:.2f} m") 
        print(f"Impact distance (without drag): {x_nd_impact:.2f} m")
        print("-----------------------------------------------------")
        print(f"Impact Speed (with drag): {v_impact:.2f} m/s") 
        print(f"Impact Speed (without drag): {v_nd_impact:.2f} m/s")
        print("-----------------------------------------------------")
        print(f"Impact Energy (with drag): {E_impact:.2f} joules") 
        print(f"Impact Energy (without drag): {E_nd_impact:.2f} joules")

    #Allows us to easily create subplots where each one shows different information.
    fig, (plt_1, plt_2, plt_3) = plt.subplots(3, 1, figsize=(8,10))
    
    #Creates a plot showing the trajectory of the projectile from its launch to the moment of impact.
    plt_1.plot(x, y, label="With Drag", color="blue")
    plt_1.plot(x_nd, y_nd, label="Without Drag", color="red", linestyle="--")
    plt_1.set_xlabel("x (meters)")
    plt_1.set_ylabel("y (meters)")
    plt_1.set_title("Projectile Motion: Drag vs No Drag")
    plt_1.grid(True)
    plt_1.legend()
    
    #Creates a plot showing the velocity of the projectile throughout its flight and its speed when it impacts the ground.
    plt_2.plot(t, v, label="With Drag", color="blue")
    plt_2.plot(t_nd, v_nd, label="Without Drag", color="red", linestyle="--")
    plt_2.set_xlabel("Time (seconds)")
    plt_2.set_ylabel("Speed (m/s)")
    plt_2.set_title("Speed vs Time")
    plt_2.grid(True)
    plt_2.legend()
    
    #Creates a plot showing the total energy (joules) of the projectile its starts with and how much energy is has at impact.
    plt_3.plot(t, E_t, label="With Drag", color="blue")
    plt_3.plot(t_nd, E_tnd, label="Without Drag", color="red", linestyle="--")
    plt_3.set_xlabel("Time (seconds)")
    plt_3.set_ylabel("Energy (joules)")
    plt_3.set_title("Total Energy vs Time")
    plt_3.grid(True)
    plt_3.legend()
    
    plt.tight_layout()

    if create_png == True: #used with above code to create a .png file containing the plots.
        fig.savefig("ProjectileMotion.png", dpi=200) 
    
    plt.show()

#Creates sliders so user can change parameters to see its effect on the projectile.
m_slider = widgets.FloatSlider(value=0.145, min=0.01, max=20, step=0.001, description="Mass (kg)")
c_slider = widgets.FloatSlider(value=0.00086, min=0.0, max=0.005, step=0.0001, description="Lumped Drag Coefficient", readout_format=".5f")
y_0_slider = widgets.FloatSlider(value=0, min=0, max=100, step=1, description="Initial Height (m)")
angle_slider = widgets.FloatSlider(value=45, min=0, max=90, step=1, description="Angle (degrees)")
v_0_slider = widgets.FloatSlider(value=30, min=1, max=100, step=1, description="Initial Speed (m/s)")

#Creates a button to reset the parameters to my intial values (baseball).
reset_button = widgets.Button(description="Reset", button_style="warning", tooltip = "Reset all sliders to inital values")

#Creates preset the user can select for various objects.
m_preset = widgets.Dropdown(options = {"Ping-Pong Ball (0.0027 kg)": 0.0027, "Baseball (0.145 kg)": 0.145, "Soccerball (0.450 kg)": 0.450, "Coconut (2.00 kg)": 2.00, "Shotput (7.26 kg)": 7.26, "Large Cannonball (14.51 kg)": 14.51}, value = 0.145, description = "Mass (kg)")
c_preset = widgets.Dropdown(options = {"Ping-Pong Ball (0.000362 kg/m)": 0.000362, "Baseball (0.00084 kg/m)": 0.00084, "Soccerball (0.00582 kg/m)": 0.00582, "Coconut (0.00701 kg/m)": 0.00701, "Shotput (0.00326 kg/m)": 0.00326, "Large Cannonball (0.00443 kg/m)": 0.00443}, value = 0.00084, description = "Lumped Drag Constant")

#Function that automatically changes the lumped drag constant to match the preset objects mass.
def c_match(correct):
    mass_change = correct['new']
    
    if mass_change == 0.145: #Baseball
        c_preset.value = 0.00084
        #c_preset.disabled = True
        m_slider.value = 0.145
        c_slider.value = 0.00084
        
    elif mass_change == 0.450: #Soccerball
        c_preset.value = 0.00582
        #c_preset.disabled = True
        m_slider.value = 0.450
        c_slider.value = 0.00582
        
    elif mass_change == 2.00: #Coconut
        c_preset.value = 0.00701
        #c_preset.disabled = True
        m_slider.value = 2.00
        c_slider.value = 0.00701
        
    elif mass_change == 7.26: #Shotput
        c_preset.value = 0.00326
        #c_preset.disabled = True
        m_slider.value = 7.26
        c_slider.value = 0.00326
        
    elif mass_change == 14.51: #Cannonball
        c_preset.value = 0.00443
        #c_preset.disabled = True
        m_slider.value = 14.51
        c_slider.value = 0.00443

    elif mass_change == 0.0027: #Ping-Pong Ball
        c_preset.value = 0.000362
        #c_preset.disabled = True
        m_slider.value = 0.0027
        c_slider.value = 0.000362
        
    else:
        #c_preset.disabled = False
        print("Use sliders")

m_preset.observe(c_match, names='value')

#Function that holds the values that the reset button, when clicked, sets each one to.
def reset_slider(b):
    m_preset.unobserve(c_match, name='value') # temporarly disable the above observer to prevent freezing.
    
    m_slider.value = 0.145
    c_slider.value = 0.00084
    y_0_slider.value = 0
    angle_slider.value = 45
    v_0_slider.value = 30
    
    m_preset.value = 0.145
    c_preset.value = 0.00084

    m_preset.observe(c_match, names='value') # re-enable observer

#Calls the above function when the reset button is clicked.
reset_button.on_click(reset_slider)

impact_values = widgets.Output()

#Stacks the reset button in a vertical column
ui = widgets.VBox([impact_values, m_slider, c_slider, y_0_slider, angle_slider, v_0_slider, m_preset, c_preset, reset_button])

#Connects the sliders to the update_simulation function.
out = widgets.interactive_output(update_simulation, {'m': m_slider, 'c': c_slider, 'y_0': y_0_slider, 'angle': angle_slider, 'v_0': v_0_slider})

#Shows everthing on the screen.
display(ui, out)
```
